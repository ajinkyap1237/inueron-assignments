```python
import numpy as np
from sklearn import cluster, datasets
import scipy.misc
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
face = scipy.misc.face()
face
```

```python
face.shape
```

```python
plt.imshow(face)
plt.show()
```

```python
x = face.reshape((-1,1))
x.shape
```

```python
n_clusters = 5

from sklearn.cluster import KMeans
k_means = KMeans(n_clusters=5, init='k-means++', n_init=4)
k_means.fit(x)
```

```python
values = k_means.cluster_centers_.squeeze()
values
```

```python
labels = k_means.labels_
labels
```

```python
face_compressed = np.choose(labels, values)
face_compressed.shape = face.shape
```

```python
plt.imshow((face_compressed*255).astype(np.uint8))
```
